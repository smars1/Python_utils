#### In the next tutorial we can see how to launch our first ```Amazon EMR Cluster on Amazon Ec2 Spot Instances```, using the Create Cluter Wizard


<h1>Step 1: </h1>

### Search AWS EMR and select the region

Go to the aws console and them seek the the ```AWS EMR in the search toolbar```. In the top right corner you can select the
 ```region```where you want to deploy the cluster, so select the region.

<h1>Step 2:</h1>

### To create our Cluster and Configure cluster software 

If you chosse you region, is moment to click on the ```create cluster``` them aws console will open the configuration window of cluster

In this Window will be possible set the requirements that you want for your cluster




<h1>Step 3:</h1>

### Configure the instance fleets

Select instance fleets under the instance group configuration section.
Instance Fleets will ```allow us to implement Instance Diversification```, which is a key best-practice when leveraging EC2 Spot Instances.

```Instance Diversification allows you to use multiple instance types```, ensuring that Amazon EMR can allocate all of the capacity needed for your cluster while automatically handling interruptions.



```Remember is recommended to choose instances with similar or proportional Vcpu and memory```



<h1>Step 4:</h1>

### Select the VPC and subnets

Also will need ```to select a VPC``` and one or more subnets where you woud like ```to deploy the Amazon EMR Cluster``` 

```We recommend choosing more than one Availability Zone.``` Your cluster will still be deployed in a single Availability Zone, however selecting multiple Availability Zones allows Amazon EMR to look across all selected Availability Zones to deploy your cluster in the Availability Zone with the most EC2 Spot Capacity to run your cluster.
